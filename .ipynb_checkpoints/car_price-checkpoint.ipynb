```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv("car_price.csv")
df
```

```python
df = df.drop(columns = ["Unnamed: 0"])
df
```

```python
# Ensure 'car_prices_in_rupee' is a string
df["car_prices_in_rupee"] = df["car_prices_in_rupee"].astype(str).str.replace(",", "").str.strip()

# Convert 'car_prices_in_rupee' to numeric (handling 'Lakh' and 'Crore')
df["car_prices_in_rupee"] = df["car_prices_in_rupee"].apply(
    lambda x: float(x.replace(" Lakh", "")) * 100000 if "Lakh" in x else
              float(x.replace(" Crore", "")) * 10000000 if "Crore" in x else 
              pd.to_numeric(x, errors="coerce")  # Convert valid numbers, else NaN
)

# Ensure 'kms_driven' is a string before processing
df["kms_driven"] = df["kms_driven"].astype(str).str.replace(" kms", "").str.replace(",", "")
df["kms_driven"] = pd.to_numeric(df["kms_driven"], errors="coerce")  # Convert to float

# Ensure 'engine' is a string before processing
df["engine"] = df["engine"].astype(str).str.replace(" cc", "")
df["engine"] = pd.to_numeric(df["engine"], errors="coerce")  # Convert to float

# Ensure 'Seats' is a string before processing
df["Seats"] = df["Seats"].astype(str).str.replace(" Seats", "")
df["Seats"] = pd.to_numeric(df["Seats"], errors="coerce")  # Convert to float


df
```

```python
le = LabelEncoder()
df["transmission"] = le.fit_transform(df["transmission"])
df["fuel_type"] = le.fit_transform(df["fuel_type"])
df["ownership"] = le.fit_transform(df["ownership"])
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
a = df["car_name"]
cv = CountVectorizer()
cc = pd.DataFrame()
c = pd.DataFrame(cv.fit_transform(df["car_name"].fillna('')).toarray())
cc = pd.concat([cc, c],axis = 1)
```
